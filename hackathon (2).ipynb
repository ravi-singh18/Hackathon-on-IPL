# 8.  Which statement holds true
*B) Teams winning tosses and electing bat first have won more number of matches

```python
win= 0
lose=0
for i in data.index:
    if data['toss_winner'][i] == data['winner'][i]:
        win+=1
    else:
        lose+=1  
```

```python
plt.figure(figsize=(12,9))
plt.title("Win vs Lose after toss winning")
plt.pie((lose,win), labels=("Lost the match","Won the match"), autopct='%1.3f%%');
```

```python
win= 0
lose=0
for i in data.index:
    if data['toss_winner'][i] != data['winner'][i]:
        win+=1
    else:
        lose+=1  
```

```python
plt.figure(figsize=(12,9))
plt.title("Win vs Lose after toss loosing")
plt.pie((lose,win), labels=("Lost the match","Won the match"), autopct='%1.3f%%');
```

# 12. Most number of wickets taken by a bowler is
*

```python
data.dismissal_kind.unique()
```

```python
data2 = data[data.dismissal_kind.notnull()].copy()
data2 = data[data.dismissal_kind != "hit wicket"]
data2 = data[data.dismissal_kind != "obstructing the field"]
data2 = data[data.dismissal_kind != "run out"]
data2 = data[data.dismissal_kind != "retired hurt"]
```

```python
bowlers = data2.bowler.unique()
bowler_wicket = pd.DataFrame(columns = ["name","wickets"])
for player in bowlers:
    tot = (data2.bowler == player).sum()
    bowler_wicket = bowler_wicket.append({"name": player,"wickets":tot},ignore_index = True)
```

```python
highest_wicket = bowler_wicket.sort_values("wickets",ascending = False).head(10)
# Now as we have made the highest wicket taker list Let's Visualize how is the trend for the top 10 wicket taker
plt.figure(figsize=(14,6))
plt.xticks(rotation=75)
plt.title("Top 10 Highest Wicket Taker")
sns.barplot(x=highest_wicket.name, y=highest_wicket.wickets).set(xlabel ="Bowler Name",ylabel = "Wickets Taken");
```
